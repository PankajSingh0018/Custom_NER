```python
## let's do some handons on kaggle dataset with json file format

import json

#dataset taken 
#'https://www.kaggle.com/datasets/finalepoch/medical-ner'

```

```python
#loading the json data 
with open('Corona2.json','r') as f:
    data = json.load(f)
    
```

```python
# printing the first example
data['examples'][0]
```

```python
# printing the keys 

data['examples'][0].keys()
```

```python
# displaying the text of the first example
data['examples'][0]['content']
```

```python
# displaying the annotation of the first example

data['examples'][0]['annotations']
```

```python
# displaying the first annotation of the first example


data['examples'][0]['annotations'][0]
```

```python
# create a training dataset 

training_data = []

for example in data['examples']:
    temp_dict = {}
    #appending the content to the dictinoary
    temp_dict['text']= example['content']

    # creating entities 
    temp_dict['entities'] = []

    for annontation in example['annotations']:
        start= annontation['start']
        end = annontation['end']
        # saving the tag_name of the entity in uppercase
        label= annontation['tag_name'].upper()

        #appending the values to the dictionary 
        temp_dict['entities'].append((start,end,label))
    
    training_data.append(temp_dict)

print(training_data)
```

```python
# entities from the training data first example
training_data[0]['entities']
```

```python
training_data[0]['text'][360:371]
```

```python
from spacy.tokens import DocBin
from tqdm import tqdm 

# loading the new spacy model
nlp= spacy.blank('en')
doc_bin = DocBin()
```

```python
from spacy.util import filter_spans

for training_example in tqdm(training_data):

    text = training_example['text']
    labels = training_example['entities']
    doc = nlp.make_doc(text)

    ents = []

    for start, end, label in labels:
        span = doc.char_span(start, end, label=label, alignment_mode='contract')
        if span is None:
            print('Skipping entity')
        else:
            ents.append(span)
    
    filtered_ents = filter_spans(ents)
    doc.ents = filtered_ents
    doc_bin.add(doc)

doc_bin.to_disk('train.spacy')
```
